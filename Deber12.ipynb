## EJERCICIO 3

```python
import numpy as np
import matplotlib.pyplot as plt

def euler_method(f, t0, y0, h, t_end):
    t_values = np.arange(t0, t_end + h, h)
    y_values = np.zeros(len(t_values))
    y_values[0] = y0

    for i in range(1, len(t_values)):
        y_values[i] = y_values[i - 1] + h * f(t_values[i - 1], y_values[i - 1])

    return t_values, y_values
```

```python
# Literal a
def f_a(t, y):
    return (y / t) - (y / t) ** 2

t_a, y_a = euler_method(f_a, 1, 1, 0.1, 2)
plt.plot(t_a, y_a, 'o-', label='Euler - Literal a')
# Configuración del gráfico
plt.xlabel('t')
plt.ylabel('y')
plt.title('Método de Euler - Ejercicio 3')
plt.legend()
plt.grid()
plt.show()
```

```python
# Literal b
def f_b(t, y):
    return 1 + (y / t) + (y / t) ** 2

t_b, y_b = euler_method(f_b, 1, 0, 0.2, 3)
plt.plot(t_b, y_b, 's-', label='Euler - Literal b')
# Configuración del gráfico
plt.xlabel('t')
plt.ylabel('y')
plt.title('Método de Euler - Ejercicio 3')
plt.legend()
plt.grid()
plt.show()
```

```python
# Literal c
def f_c(t, y):
    return -(y + 1) * (y + 3)

t_c, y_c = euler_method(f_c, 0, -2, 0.2, 2)
plt.plot(t_c, y_c, 'd-', label='Euler - Literal c')
# Configuración del gráfico
plt.xlabel('t')
plt.ylabel('y')
plt.title('Método de Euler - Ejercicio 3')
plt.legend()
plt.grid()
plt.show()
```

```python
# Literal d
def f_d(t, y):
    return -5*y + 5*t**2 + 2*t

t_d, y_d = euler_method(f_d, 0, 1/3, 0.1, 1)
plt.plot(t_d, y_d, 'x-', label='Euler - Literal d')
# Configuración del gráfico
plt.xlabel('t')
plt.ylabel('y')
plt.title('Método de Euler - Ejercicio 3')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def euler_method(f, t0, y0, h, t_end):
    t_values = np.arange(t0, t_end + h, h)
    y_values = np.zeros(len(t_values))
    y_values[0] = y0

    for i in range(1, len(t_values)):
        y_values[i] = y_values[i - 1] + h * f(t_values[i - 1], y_values[i - 1])

    return t_values, y_values

# Soluciones exactas
def exact_a(t):
    return t / (1 + np.log(t))

def exact_b(t):
    return t * np.tan(np.log(t))

def exact_c(t):
    return -3 + 2 / (1 + np.exp(-2*t))

def exact_d(t):
    return t**2 + (1/3) * np.exp(-5*t)

# Cálculo de errores
def compute_error(approx_values, exact_func, t_values):
    exact_values = exact_func(t_values)
    error = np.abs(exact_values - approx_values)
    return np.round(error, 3)
```

```python
# Literal a
def f_a(t, y):
    return (y / t) - (y / t) ** 2

t_a, y_a = euler_method(f_a, 1, 1, 0.1, 2)
error_a = compute_error(y_a, exact_a, t_a)
plt.plot(t_a, error_a, 'o-', label='Error - Literal a')
# Configuración del gráfico
plt.xlabel('t')
plt.ylabel('Error')
plt.title('Error del Método de Euler vs Solución Exacta - Ejercicio 4')
plt.legend()
plt.grid()
plt.show()
```

```python
# Literal b
def f_b(t, y):
    return 1 + (y / t) + (y / t) ** 2

t_b, y_b = euler_method(f_b, 1, 0, 0.2, 3)
error_b = compute_error(y_b, exact_b, t_b)
plt.plot(t_b, error_b, 's-', label='Error - Literal b')
# Configuración del gráfico
plt.xlabel('t')
plt.ylabel('Error')
plt.title('Error del Método de Euler vs Solución Exacta - Ejercicio 4')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

def euler_method(f, t0, y0, h, t_end):
    t_values = np.arange(t0, t_end + h, h)
    y_values = np.zeros(len(t_values))
    y_values[0] = y0

    for i in range(1, len(t_values)):
        y_values[i] = y_values[i - 1] + h * f(t_values[i - 1], y_values[i - 1])

    return t_values, y_values

# Función para graficar comparaciones
def plot_comparison(t_values, y_values, exact_func, query_points, label):
    interp_func = interp1d(t_values, y_values, kind='linear', fill_value='extrapolate')
    interpolated_values = interp_func(query_points)
    query_points = np.array(query_points)  # Asegurar que query_points es un array
    exact_values = exact_func(query_points)

    plt.figure(figsize=(6, 4))
    plt.plot(t_values, y_values, 'o-', label='Euler')
    plt.plot(query_points, interpolated_values, 's', label='Interpolado', markersize=8)
    plt.plot(query_points, exact_values, 'x', label='Exacto', markersize=8, color='red')
    plt.xlabel('t')
    plt.ylabel('y')
    plt.title(f'Comparación - {label}')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
# Literal a
def exact_a(t):
    return t / (1 + np.log(t))

def f_a(t, y):
    return (y / t) - (y / t) ** 2

t_a, y_a = euler_method(f_a, 1, 1, 0.1, 2)
query_points_a = [0.25, 0.93]
plot_comparison(t_a, y_a, exact_a, query_points_a, "Literal a")
```

```python
# Literal b
def exact_b(t):
    return t * np.tan(np.log(t))

def f_b(t, y):
    return 1 + (y / t) + (y / t) ** 2

t_b, y_b = euler_method(f_b, 1, 0, 0.2, 3)
query_points_b = [1.25, 1.93]
plot_comparison(t_b, y_b, exact_b, query_points_b, "Literal b")
```

```python
# Literal c
def exact_c(t):
    t = np.array(t)
    return -3 + 2 / (1 + np.exp(-2*t))

def f_c(t, y):
    return -(y + 1) * (y + 3)

t_c, y_c = euler_method(f_c, 0, -2, 0.2, 2)
query_points_c = [2.10, 2.75]
plot_comparison(t_c, y_c, exact_c, query_points_c, "Literal c")
```

```python
# Literal d
def exact_d(t):
    t = np.array(t)
    return t**2 + (1/3) * np.exp(-5*t)

def f_d(t, y):
    return -5*y + 5*t**2 + 2*t

t_d, y_d = euler_method(f_d, 0, 1/3, 0.1, 1)
query_points_d = [0.54, 0.94]
plot_comparison(t_d, y_d, exact_d, query_points_d, "Literal d")
```
