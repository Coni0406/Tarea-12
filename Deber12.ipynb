```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

def euler_method(f, t0, y0, h, t_end):
    t_values = np.arange(t0, t_end + h, h)
    y_values = np.zeros(len(t_values))
    y_values[0] = y0

    for i in range(1, len(t_values)):
        y_values[i] = y_values[i - 1] + h * f(t_values[i - 1], y_values[i - 1])

    return t_values, y_values

# Función para graficar comparaciones
def plot_comparison(t_values, y_values, exact_func, query_points, label):
    interp_func = interp1d(t_values, y_values, kind='linear', fill_value='extrapolate')
    interpolated_values = interp_func(query_points)
    query_points = np.array(query_points)  # Asegurar que query_points es un array
    exact_values = exact_func(query_points)

    plt.figure(figsize=(6, 4))
    plt.plot(t_values, y_values, 'o-', label='Euler')
    plt.plot(query_points, interpolated_values, 's', label='Interpolado', markersize=8)
    plt.plot(query_points, exact_values, 'x', label='Exacto', markersize=8, color='red')
    plt.xlabel('t')
    plt.ylabel('y')
    plt.title(f'Comparación - {label}')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
# Literal a
def exact_a(t):
    return t / (1 + np.log(t))

def f_a(t, y):
    return (y / t) - (y / t) ** 2

t_a, y_a = euler_method(f_a, 1, 1, 0.1, 2)
query_points_a = [0.25, 0.93]
plot_comparison(t_a, y_a, exact_a, query_points_a, "Literal a")
```

```python
# Literal b
def exact_b(t):
    return t * np.tan(np.log(t))

def f_b(t, y):
    return 1 + (y / t) + (y / t) ** 2

t_b, y_b = euler_method(f_b, 1, 0, 0.2, 3)
query_points_b = [1.25, 1.93]
plot_comparison(t_b, y_b, exact_b, query_points_b, "Literal b")
```

```python
# Literal c
def exact_c(t):
    t = np.array(t)
    return -3 + 2 / (1 + np.exp(-2*t))

def f_c(t, y):
    return -(y + 1) * (y + 3)

t_c, y_c = euler_method(f_c, 0, -2, 0.2, 2)
query_points_c = [2.10, 2.75]
plot_comparison(t_c, y_c, exact_c, query_points_c, "Literal c")
```

```python
# Literal d
def exact_d(t):
    t = np.array(t)
    return t**2 + (1/3) * np.exp(-5*t)

def f_d(t, y):
    return -5*y + 5*t**2 + 2*t

t_d, y_d = euler_method(f_d, 0, 1/3, 0.1, 1)
query_points_d = [0.54, 0.94]
plot_comparison(t_d, y_d, exact_d, query_points_d, "Literal d")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def taylor2_method(f, df, t0, y0, h, t_end):
    t_values = np.arange(t0, t_end + h, h)
    y_values = np.zeros(len(t_values))
    y_values[0] = y0

    for i in range(1, len(t_values)):
        t, y = t_values[i - 1], y_values[i - 1]
        y_values[i] = y + h * f(t, y) + (h**2 / 2) * df(t, y)

    return t_values, y_values
```

```python
# Literal a
def f_a(t, y):
    return t * np.exp(3*t) - 2*y

def df_a(t, y):
    return (np.exp(3*t) + 3*t*np.exp(3*t)) - 2*f_a(t, y)

t_a, y_a = taylor2_method(f_a, df_a, 0, 0, 0.5, 1)
plt.figure()
plt.plot(t_a, y_a, 'o-', label='Taylor 2 - Literal a')
plt.xlabel('t')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.title('Literal a')
plt.show()
```

```python
# Literal b
def f_b(t, y):
    return 1 + (t - y) ** 2

def df_b(t, y):
    return 2 * (t - y) * (1 - f_b(t, y))

t_b, y_b = taylor2_method(f_b, df_b, 2, 1, 0.5, 3)
plt.figure()
plt.plot(t_b, y_b, 's-', label='Taylor 2 - Literal b')
plt.xlabel('t')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.title('Literal b')
plt.show()
```

```python
# Literal c
def f_c(t, y):
    return 1 + y / t

def df_c(t, y):
    return (1 / t) * (1 + f_c(t, y))

t_c, y_c = taylor2_method(f_c, df_c, 1, 2, 0.25, 2)
plt.figure()
plt.plot(t_c, y_c, 'd-', label='Taylor 2 - Literal c')
plt.xlabel('t')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.title('Literal c')
plt.show()
```

```python
# Literal d
def f_d(t, y):
    return np.cos(2*t) + np.sin(3*t)

def df_d(t, y):
    return -2*np.sin(2*t) + 3*np.cos(3*t)

t_d, y_d = taylor2_method(f_d, df_d, 0, 1, 0.25, 1)
plt.figure()
plt.plot(t_d, y_d, 'x-', label='Taylor 2 - Literal d')
plt.xlabel('t')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.title('Literal d')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def taylor4_method(f, df, d2f, d3f, t0, y0, h, t_end):
    t_values = np.arange(t0, t_end + h, h)
    y_values = np.zeros(len(t_values))
    y_values[0] = y0

    for i in range(1, len(t_values)):
        t, y = t_values[i - 1], y_values[i - 1]
        y_values[i] = y + h * f(t, y) + (h**2 / 2) * df(t, y) + (h**3 / 6) * d2f(t, y) + (h**4 / 24) * d3f(t, y)

    return t_values, y_values
```

```python
# Literal a
def f_a(t, y):
    return t * np.exp(3*t) - 2*y

def df_a(t, y):
    return (np.exp(3*t) + 3*t*np.exp(3*t)) - 2*f_a(t, y)

def d2f_a(t, y):
    return (6*np.exp(3*t) + 9*t*np.exp(3*t)) - 2*df_a(t, y)

def d3f_a(t, y):
    return (27*np.exp(3*t) + 27*t*np.exp(3*t)) - 2*d2f_a(t, y)

t_a4, y_a4 = taylor4_method(f_a, df_a, d2f_a, d3f_a, 0, 0, 0.5, 1)
plt.plot(t_a4, y_a4, 'o-', label='Taylor 4 - Literal a')

# Configuración del gráfico
plt.xlabel('t')
plt.ylabel('y')
plt.title('Método de Taylor de Orden 4')
plt.legend()
plt.grid()
plt.show()
```

```python
# Literal b
def f_b(t, y):
    return 1 + (t - y) ** 2

def df_b(t, y):
    return 2 * (t - y) * (1 - f_b(t, y))

def d2f_b(t, y):
    return 2 * (1 - f_b(t, y))**2 - 2*f_b(t, y) * df_b(t, y)

def d3f_b(t, y):
    return -6 * df_b(t, y) * (1 - f_b(t, y)) - 2*f_b(t, y) * d2f_b(t, y)

t_b4, y_b4 = taylor4_method(f_b, df_b, d2f_b, d3f_b, 2, 1, 0.5, 3)
plt.plot(t_b4, y_b4, 's-', label='Taylor 4 - Literal b')

# Configuración del gráfico
plt.xlabel('t')
plt.ylabel('y')
plt.title('Método de Taylor de Orden 4')
plt.legend()
plt.grid()
plt.show()
```

```python
# Literal c
def f_c(t, y):
    return 1 + y / t

def df_c(t, y):
    return (1 / t) * (1 + f_c(t, y))

def d2f_c(t, y):
    return (-1 / t**2) * (1 + f_c(t, y)) + (1/t) * df_c(t, y)

def d3f_c(t, y):
    return (2 / t**3) * (1 + f_c(t, y)) - (2 / t**2) * df_c(t, y) + (1 / t) * d2f_c(t, y)

t_c4, y_c4 = taylor4_method(f_c, df_c, d2f_c, d3f_c, 1, 2, 0.25, 2)
plt.plot(t_c4, y_c4, 'd-', label='Taylor 4 - Literal c')

# Configuración del gráfico
plt.xlabel('t')
plt.ylabel('y')
plt.title('Método de Taylor de Orden 4')
plt.legend()
plt.grid()
plt.show()
```
